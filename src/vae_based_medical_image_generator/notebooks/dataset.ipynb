# Dataset

```python
import medmnist

print(f"MedMNIST v{medmnist.__version__} @ {medmnist.HOMEPAGE}")
```

```python
from vae_based_medical_image_generator.data import dataset

datasets = {dataset_name: dataset.load_dataset(dataset_name=dataset_name, split="train") for dataset_name in medmnist.INFO.keys() if "3d" not in dataset_name}
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 4, figsize=(10, 8))
axs = axs.flatten()
for idx, (dataset_name, dataset) in enumerate(datasets.items()):
    axs[idx].axis('off')
    axs[idx].set_title(dataset_name)
    axs[idx].imshow(dataset.montage(length=20), cmap='gray')

plt.show()
```
